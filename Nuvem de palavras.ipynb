```python
##Import das bibliotecas
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import warnings
```

```python
##conjunto de palavras fornceidas para WordCloud
palavras = ("test","git","Python", "github")
print(palavras)

words = ' '.join(palavras)

##Exclusão de alguma palavra caso necessário
words_clean = " ".join([word for word in words.split()
                            if 'https' not in word
                            ])
```

```python
import warnings
import matplotlib.pyplot as plt

##Suprimindo avisos
warnings.simplefilter('ignore')

wc = WordCloud(min_font_size=10, 
               max_font_size=300, 
               background_color='black', 
               mode="RGB",
               width=2000, 
               height=1000,
               normalize_plurals= True).generate(words_clean)

plt.imshow(wc, interpolation="lanczos")
plt.axis("off")
plt.savefig('words_cloud_pt.png', dpi=300)
plt.show()
```
